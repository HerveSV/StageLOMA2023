```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
def duffing_deriv(t, X, gamma, epsilon, f0, omega_d):
    x, y = X
    return [y, -gamma*y -x -epsilon**3*x + f0*np.cos(omega_d*t)]

EPSILON = 1#0.1#0#3
F0 = 1
GAMMA = 0.01

OMEGA_DRIVE = 5#2*np.pi


t_start = 0.0
t_stop = 10000

DURATION = t_stop
SAMPLE_RATE = 1000
SAMPLES = int(DURATION*SAMPLE_RATE)
t = np.linspace(t_start, t_stop, SAMPLES)

x0 = [2, 0]

sol = solve_ivp(duffing_deriv,
                [t_start, t_stop],
                x0,
                t_eval=t,
                args=[GAMMA, EPSILON, F0, OMEGA_DRIVE],
                method="LSODA",
                dense_output=True)

x, y = sol.y

plt.plot(x, y)
plt.grid()
plt.axis('equal')
plt.show()

plt.plot(t, x)
plt.grid()
#plt.xlim(0, 6)
#plt.ylim(-0.2, 0.2)
plt.show()

plt.plot(t, y)
plt.grid()
#plt.ylim(-0.2, 0.2)
plt.show()
```
