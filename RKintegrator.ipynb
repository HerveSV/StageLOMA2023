```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
import random
```

```python
# dx/dt = y
# dy/dt = -x - eps(x^2 - 1)y

def vdp(t, x, eps):
    X, Y = x
    return [Y, -eps*(X*X - 1)*Y - X]

EPSILON = 0.1
EPSILON_LARGE = 10

t_start = 0.0
t_stop = 200.0
t = np.linspace(t_start, t_stop, 100000)





'''sol = solve_ivp(vdp, 
                [t_start, t_stop],
                [x0, v0],
                args=[EPSILON],
                method="LSODA",
                dense_output=True)
'''
```

```python
X0 = [random.uniform(0, 3) for i in range(5)]#[0.1, 0.5, 3.5, 5]
Y0 = [random.uniform(0, 3) for i in range(5)]#[0, 0, 0, 0]

```

```python
sols = [[] for e in X0]

for i in range(len(X0)):
    sols[i] = solve_ivp(vdp, 
                [t_start, t_stop],
                [X0[i], Y0[i]],
                args=[EPSILON],
                method="LSODA",
                dense_output=True,
                t_eval=t)
    
sols_large = [[] for e in X0]
for i in range(len(X0)):
    sols_large[i] = solve_ivp(vdp, 
                [t_start, t_stop],
                [X0[i], Y0[i]],
                args=[EPSILON_LARGE],
                method="LSODA",
                dense_output=True,
                t_eval=t)
#t = sol.t
#X, V = sol.y

'''
plt.plot(t, X)
plt.show()
plt.plot(t, V)
plt.show()
'''



for i in range(len(X0)):
    t = sols[i].t
    X, Y = sols[i].y

    plt.plot(X, Y, alpha=0.7)
    if(i == 0):
        plt.scatter(X0[i], Y0[i], color="r", label="Conditions initiales", zorder=5, alpha=0.9)
    else:
        plt.scatter(X0[i], Y0[i], color="r", zorder=5, alpha=0.9)
    plt.axis("equal")

plt.xlabel("$x$")
plt.ylabel("$\dot{x}$")
plt.title(f"$\epsilon = ${EPSILON}")
plt.legend()
#plt.savefig('figs/vdp/vanderpol_small.png', dpi=300, format='png', transparent=False, bbox_inches='tight')
plt.show()

plt.plot(t, sols[2].y[0])
plt.title(f"$\epsilon = ${EPSILON}")
plt.ylabel("$x$")
plt.xlabel("$t$")
#plt.savefig('figs/vdp/vanderpol_small_x.png', dpi=300, format='png', transparent=False, bbox_inches='tight')
plt.show()





for i in range(len(X0)):
    t = sols_large[i].t
    X, Y = sols_large[i].y

    plt.plot(X, Y, alpha=0.7)
    if(i == 0):
        plt.scatter(X0[i], Y0[i], color="r", label="Conditions initiales", zorder=5, alpha=0.9)
    else:
        plt.scatter(X0[i], Y0[i], color="r", zorder=10, alpha=0.9)
    
plt.xlim(-4, 4)
plt.xlabel("$x$")
plt.ylabel("$\dot{x}$")
plt.title(f"$\epsilon = ${EPSILON_LARGE}")
plt.legend()
#plt.savefig('figs/vdp/vanderpol_large.png', dpi=300, format='png', transparent=False, bbox_inches='tight')
plt.show()

plt.plot(t, sols_large[4].y[0])
plt.title(f"$\epsilon = ${EPSILON_LARGE}")
plt.ylabel("$x$")
plt.xlabel("$t$")
#plt.savefig('figs/vdp/vanderpol_large_x.png', dpi=300, format='png', transparent=False, bbox_inches='tight')
plt.show()





```

```python
def x_lindsted_approx(t, eps):
    w = 1 - 1/16*eps**2
    tau = w*t
    return 2*np.cos(tau) + eps*(3/4*np.sin(tau) - 1/4*np.sin(3*tau)) + eps**2*(3/16*np.cos(3*tau) - 5/96*np.cos(5*tau) - 41/304*np.cos(tau))

def transient_approx(t, eps, x0):
    A = x0**2 / (4 - x0**2)
    r = 2*np.sqrt(A/(A + np.exp(-eps*t)))
    return r*np.cos(t)



x_approx1 = x_lindsted_approx(t, EPSILON)
plt.plot(t, x_approx1, label="Approx. Linsted", alpha=0.8, color="tab:blue")
plt.plot(t, x, alpha=0.5, label="Solution numérique", color="r")
plt.xlabel("$t$")
plt.ylabel("$x$")
plt.legend(loc='center')
#plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.13),
#          fancybox=True, shadow=True, ncol=5)
#plt.title(f"Approximation du cyle limite, $x(0)=${x0}, $\dot x(0) = ${y0}, $\epsilon = ${EPSILON}")
plt.savefig('figs/vdp/approx_linsted.png', dpi=300, format='png', transparent=False, bbox_inches='tight')
plt.show()

x_approx2 = transient_approx(t, EPSILON, x0)

plt.plot(t, x_approx2, label="Approx. moyennement", alpha=0.8, color="tab:blue")
plt.plot(t, x, alpha=0.5, label="Solution numérique", color="r")
plt.xlabel("$t$")
plt.ylabel("$x$")
plt.legend(loc='center')
#plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.13),
#          fancybox=True, shadow=True, ncol=5)
#plt.title(f"Approximation du régime transient, $x(0)=${x0}, $\dot x(0) = ${y0}, $\epsilon = ${EPSILON}")
plt.savefig('figs/vdp/approx_transient.png', dpi=300, format='png', transparent=False, bbox_inches='tight')
plt.show()
```
