```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
eps = 0.01
f0 = 1.4#1.243
f02 = f0**2



q2 = np.linspace(eps, f02, 10000)
w_plus = q2 + np.sqrt((f02/q2) - 1)
w_min = q2 - np.sqrt((f02/q2) - 1)
w = w_plus + w_min

plt.plot(q2, w_min, color="blue")
plt.plot(q2, w_plus, color="purple")
plt.show
```

```python
def wplus_gradient(x, f2):
    return 1 - 0.5/(np.sqrt(x**3/f2 - x**4/f2**2))

def cfd_grad(T, X):
    dX = np.zeros(len(X)-2)
    for i in range(len(dX)):
        dX[i] = (X[i+2]-X[i])/(T[i+2]-T[i])
    return dX

F0 = np.linspace(1.2, 1.5, 1000)
F02  = F0**2

x0 = 3/4 * F02
grads = wplus_gradient(x0, F02)
#grads = cfd

smallest = 100000000
small_index = 0
for i in range(len(grads)):
    if np.abs(grads[i]) < smallest:
        smallest = np.abs(grads[i])
        small_index = i

print(f"smallest gradient of {smallest} for F0 value of {F0[i]:.6}")
print(F0[i])
```
