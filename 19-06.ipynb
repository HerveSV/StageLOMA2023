```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w0 = 3
gamma = 0.2
W = np.linspace(0, 8, 15000)
```

```python



Rho2 = 1/((w0**2 - W**2)**2 + W**2*gamma**2)

rho2max = 1/(w0**2*gamma**2)




plt.plot(W, Rho2, zorder=0)
plt.scatter(w0, rho2max, color="r", zorder=1, s=25, label="$\omega = \omega_0$")


plt.xlim(1, 5)
plt.ylim(-0.3, 12)
plt.legend()
plt.title(f"$\\rho^2$ en fonction de $\omega$ ($\omega_0 = {w0}, \gamma = {gamma}$)")
plt.ylabel("$\\rho^2$")
plt.xlabel("$\omega$")

plt.savefig('figs/rho_plot.png', dpi=300, format='png', transparent=False)
plt.show()


gamma2 = 0.01
Rho2_2 = 1/((w0**2 - W**2)**2 + W**2*gamma2)
rho2max_2 = 1/(w0**2*gamma2)


plt.plot(W, Rho2_2, zorder=0)
plt.scatter(w0, rho2max_2, color="r", zorder=1, s=25, label="$\omega = \omega_0$")

plt.xlim(1, 5)
plt.ylim(-0.3, 25)
plt.legend()
plt.title(f"$\\rho^2$ en fonction de $\omega$ ($\omega_0 = {w0}, \gamma = {gamma2}$)")
plt.ylabel("$\\rho^2$")
plt.xlabel("$\omega$")
plt.savefig('figs/rho_plot_smallgamma.png', dpi=300, format='png', transparent=False)
plt.show()


fig, axes = plt.subplots(1,2, sharey=True)
axes[0].plot(W, Rho2, zorder=0)
axes[0].scatter(w0, rho2max, color="r", zorder=1, s=25, label="$\omega = \omega_0$")
axes[0].set_xlim(1, 5)
axes[0].set_ylim(-0.3, 12)
axes[0].legend()
axes[0].set_title(f"($\omega_0 = {w0}, \gamma = {gamma}$)")
axes[0].set_ylabel("$\\rho^2$")
axes[0].set_xlabel("$\omega$")

axes[1].plot(W, Rho2_2, zorder=0)
axes[1].scatter(w0, rho2max_2, color="r", zorder=1, s=25, label="$\omega = \omega_0$")
axes[1].set_xlim(1, 5)
axes[1].set_ylim(-0.3, 12)
axes[1].legend()
axes[1].set_title(f"($\omega_0 = {w0}, \gamma = {gamma2}$)")
axes[1].set_ylabel("$\\rho^2$")
axes[1].set_xlabel("$\omega$")

fig.suptitle("$\\rho^2$ en fonction de $\omega$ ")
plt.savefig('figs/rho2_plot_contrast.png', dpi=300, format='png', transparent=False)

plt.show()

Rho = np.sqrt(Rho2)
Rho_2 = np.sqrt(Rho2_2)
rho_max = np.sqrt(rho2max)
rho_max_2 = np.sqrt(rho2max_2)

fig, axes = plt.subplots(1,2, sharey=True)
axes[0].plot(W, Rho, zorder=0)
axes[0].scatter(w0, rho_max, color="r", zorder=1, s=25, label="$\omega = \omega_0$")
axes[0].set_xlim(1, 5)
axes[0].set_ylim(-0.3, 4)
axes[0].legend()
axes[0].set_title(f"$\omega_0 = {w0}, \gamma = {gamma}$")
axes[0].set_ylabel("$\\rho$")
axes[0].set_xlabel("$\omega$")

axes[1].plot(W, Rho_2, zorder=0)
axes[1].scatter(w0, rho_max_2, color="r", zorder=1, s=25, label="$\omega = \omega_0$")
axes[1].set_xlim(1, 5)
axes[1].set_ylim(-0.3, 4)
axes[1].legend()
axes[1].set_title(f"$\omega_0 = {w0}, \gamma = {gamma2}$")
axes[1].set_ylabel("$\\rho$")
axes[1].set_xlabel("$\omega$")

plt.savefig('figs/rho_plot_contrast.png', dpi=300, format='png', transparent=False)

plt.show()

'''
Theta = np.rad2deg(np.arctan2((w0**2 - W**2), -W*gamma))
plt.plot(W, Theta)
plt.show()

'''
```

```python
R = 1/(w0**2 - W**2 + W*gamma*1j)
x = [e.real for e in R]
y = [e.imag for e in R]
Rmax = 1/(+ w0*gamma*1j)
xmax = Rmax.real
ymax = Rmax.imag

delta_w = 0.00076*w0#0.00230
Rbefore = 1/(w0**2 - (w0-delta_w)**2 + (w0-delta_w)*gamma*1j)
Rafter = 1/(w0**2 - (w0+delta_w)**2 + (w0+delta_w)*gamma*1j)

plt.scatter(x, y)
plt.scatter(xmax, ymax, label="$\omega = \omega_0$", color="r")
plt.scatter(Rbefore.real, Rbefore.imag, label="$\omega < \omega_0$")
plt.scatter(Rafter.real, Rafter.imag, label="$\omega > \omega_0$")

plt.title("Valeurs complexes de $R$ en fonction de $\omega$")
plt.legend(loc="center")
plt.ylabel("Axe imaginaire")
plt.xlabel("Axe relle")
plt.savefig('figs/R_valeurs.png', dpi=300, format='png', transparent=False)
plt.show()


theta = np.angle(R)
plt.plot(W, theta)
plt.xlim(0,6)
plt.show()

```

```python

t = np.linspace(0, 10, 10000)




gamma = 3
w0 = 12
f0 = 0.3

wd = w0

Rp = 1/(w0**2 - wd**2 + wd*gamma*1j)

A = 20
xh = np.exp(-gamma/2*t)*np.cos(np.sqrt(w0**2 - gamma**2/4)*t)
xp = A*np.abs(Rp)*f0*np.cos(wd*t + np.angle(Rp))
print(np.abs(Rp))

plt.plot(t, xh, alpha=0.5, color="blue", label="solution homogne")
#plt.plot(t, xp)
plt.plot(t, xh+xp, color="orange", label="solution gnrale")
plt.ylim(-1, 1)
plt.legend()
plt.title("Solutions pour $x$, ($\omega_0 > \gamma/2$)")

#plt.savefig('figs/solutions2.png', dpi=300, format='png', transparent=False)
plt.show()

# deviation from w0, signal is less strong
wd2 = w0 + 3
Rp2 = 1/(w0**2 - wd2**2 + wd2*gamma*1j)
xp = A*np.abs(Rp2)*f0*np.cos(wd2*t + np.angle(Rp2))
plt.plot(t, xh, alpha=0.5, color="blue", label="solution homogne")
#plt.plot(t, xp)
plt.plot(t, xh+xp, color="orange", label="solution gnrale")
plt.ylim(-1, 1)
plt.legend()
plt.title("Solutions pour $x$, ($\omega_0 > \gamma/2$)")
plt.show()


vh = np.sqrt(w0**2 - gamma**2/4)*(gamma/2)*np.exp(-gamma/2*t)*np.sin(np.sqrt(w0**2 - gamma**2/4)*t)
vp = -wd*A*np.abs(Rp)*f0*np.sin(wd*t + np.angle(Rp))

v = vp+vh
x = xp+xh

plt.plot(x, v)
plt.show()

```
